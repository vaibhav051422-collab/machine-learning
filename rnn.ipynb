```python
from keras import Sequential
import tensorflow
from keras.layers import Dense ,SimpleRNN
```

```python
model=Sequential()
model.add(SimpleRNN(3,input_shape=(4,5)))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```
