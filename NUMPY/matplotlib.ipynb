```python
# Real data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]

oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

# Plot
plt.plot(years,oscar_revenue)
plt.title("oscars movie revenue")
plt.xlabel("years")
plt.ylabel("revenue")
```

```python
nonoscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years = [2003,2004,2006,2019,2011]
nonoscar_revenue = [1005, 180, 627, 143, 233]  # in $M
plt.plot(years,oscar_revenue,label="oscar winners")
plt.plot(years,nonoscar_revenue, "o--r",label="non oscar winners")
plt.xlabel("years")
plt.ylabel("revenue")
plt.title("oscars movie revenue vs non oscar")
plt.tight_layout()
plt.legend()
plt.savefig("pic.png")



```

```python
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]

oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

# Plot
plt.bar(years,oscar_revenue)
plt.title("oscars movie revenue")
plt.xlabel("years")
plt.ylabel("revenue")
for i in range (len(years)):
   plt.text(years[i],oscar_revenue[i],str(oscar_revenue[i]),ha='center')


```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
scores=np.random.normal(70,10,100)
plt.hist(scores,color="skyblue",bins=20)
plt.axvline(
    x=33,
    label='passing marks (33)',
    color="red",
    linewidth="2"
)
```
